```python
import folium
from folium.plugins import HeatMap
from folium import plugins
import pandas as pd
```

```python
dados = pd.read_csv("states.csv")
```

```python
dados
```

```python
dados.info
```

```python
baseMap = folium.Map(
    widht = '100%',
    height = '100%',
    location = [-15.788497, -47.899873],
    zomm_start = 4
)
```

```python
baseMap
```

```python
coordenadas = dados [['Latitude', 'Longitude', 'GDP']].values.tolist()
```

```python
coordenadas
```

```python
HeatMap(coordenadas, radius = 15).add_to(baseMap)
```

```python
baseMap
```

```python
for i in range(0, len(dados)):
    folium.Circle(
        location=[dados.iloc[i]['Latitude'], dados.iloc[i]['Longitude']],
        color='#000000',
        fill='#00A103',
        tooltip='<li><bold> Estado : ' + str(dados.iloc[i]['State']) +
                '<li><bold> População : ' + str(dados.iloc[i]['Population']) +
                '<li><bold> GDP : ' + str(dados.iloc[i]['GDP']),
        radius=10
    ).add_to(baseMap)
```
